```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv("covid_pa_confirmed_data_7d.csv")
data.value = data.value.astype(int)
data
```

```python
def bootstrap(data, space ,overlap):
    bootstrap_data = []
    i = 0
    while i < len(data):
        bootstrap_data.append(data[i+space-overlap:i+space].mean())
        i = i+space
    return bootstrap_data
bootstrap_data = bootstrap(data.value, space = 2, overlap = 1)
print(len(data.value),len(bootstrap_data))


```
